## 1\. Pinguins 

Neste exercício, irei utilizar uma base de dados com informações sobre penguins. A idéia é preparar a base de dados para prever a espécie do penguin (variável resposta) baseado em suas características físicas e geográficas (variáveis preditivas).

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset('penguins')
```

```python
data.head()
```

### **1.1. Valores nulos** 

A base de dados possui valores faltantes, utilize os conceitos da aula para trata-los.

### **1.2. Variáveis numéricas** 

Identifique as variáveis numéricas e crie uma nova coluna **padronizando** seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_std*".

```python
#data['bill_length_mm'].fillna(data['bill_length_mm'].mean(), inplace=True)
#data['bill_depth_mm'].fillna(data['bill_depth_mm'].mean(), inplace=True)
#data['flipper_length_mm'].fillna(data['flipper_length_mm'].mean(), inplace=True)
#data['body_mass_g'].fillna(data['body_mass_g'].mean(), inplace=True)
```

```python
# Identificar colunas numéricas
numeric_cols = data.select_dtypes(include=np.number).columns

# Verificar valores faltantes nas colunas numéricas
missing_values = data[numeric_cols].isnull().sum()

# Preencher valores faltantes com a média
for col in numeric_cols:
    col_name_std = col + "_std"
    mean_value = data[col].mean()
    data[col_name_std] = data[col].fillna(mean_value)

# Verificar as primeiras linhas do dataframe com as novas colunas com valores preenchidos
data.head()
```

```python
print(data.isnull().sum())
```

### **1.3. Variáveis categóricas** 

Identifique as variáveis categóricas nominais e ordinais, crie uma nova coluna aplicando a técnica correta de conversão a seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_nom*" ou "*_ord*".

```python
# Identificar colunas categóricas
categorical_cols = data.select_dtypes(include=['object']).columns
```

```python
# Conversão das variáveis categóricas nominais
for col in ['sex']:
    col_name_nom = col + "_nom"
    data[col_name_nom] = pd.Categorical(data[col]).codes

# Verificar as primeiras linhas do dataframe com as novas colunas de variáveis categóricas nominais
data.head()
```

### **1.4. Limpeza** 

Descarte as colunas originais e mantenha apenas a variável resposta e as variáveis preditivas com o sufixo *_std*", *_nom*" e "*_ord*". 

```python
# Criar lista com os sufixos das colunas que queremos manter
sufixos = ['_std', '_nom']

# Criar lista com as colunas que queremos manter
colunas_mantidas = [col for col in data.columns if any(col.endswith(sufixo) for sufixo in sufixos)]

# Adicionar a variável resposta à lista de colunas mantidas
colunas_mantidas.append('species')
colunas_mantidas.append('island')
# Criar um novo dataframe com as colunas desejadas
data_final = data[colunas_mantidas]

# Verificar as primeiras linhas do novo dataframe
data_final.head()
```
